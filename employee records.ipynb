2. SALARY ANALYSIS

a. The department with the highest average salary

SELECT department, ROUND (AVG(salary), 2) AS avg_salary
FROM employee_records
GROUP BY department
ORDER BY AVG(salary) DESC;

b. The positions with the highest salaries

SELECT position, ROUND(MAX(salary), 2) AS highest_salary
FROM employee_records
GROUP BY position
ORDER BY MAX(salary) DESC;

c. To check whether salary increases with experience -

SELECT
    CASE
        WHEN experience < 5 THEN '0-4 Years'
        WHEN experience BETWEEN 5 AND 9 THEN '5-9 Years'
        WHEN experience BETWEEN 10 AND 14 THEN '10-14 Years'
        WHEN experience BETWEEN 15 AND 19 THEN '15-19 Years'
        ELSE '20+ Years'
    END AS experience_group,
    ROUND(AVG(salary), 2) AS average_salary,
    COUNT(*) AS employees
FROM employee_records
GROUP BY experience_group
ORDER BY employees DESC;

FINDING: Salary does not increase with experience.

d. To check salary differences among education levels -

SELECT
    education_level,
    MIN(salary) AS lowest_salary,
    MAX(salary) AS highest_salary,
    ROUND(AVG(salary), 2) AS average_salary
FROM employee_records
GROUP BY education_level
ORDER BY AVG(salary) DESC;

FINDING: PhD holders are the highest paid and employees holding only a Diploma are the lowest paid.

e. To check whether salary differences exist between genders in the same department -

SELECT
    department,
    ROUND(AVG(CASE WHEN gender = 'M' THEN salary END), 2) AS male_avg_salary,
    ROUND(AVG(CASE WHEN gender = 'F' THEN salary END), 2) AS female_avg_salary,
    ROUND(
        AVG(CASE WHEN gender = 'M' THEN salary END) -
        AVG(CASE WHEN gender = 'F' THEN salary END),
        2
    ) AS salary_difference
FROM employee_records
GROUP BY department;

Disclaimer: Average salaries differ between genders in some departments. However, this analysis does not account for other factors such as job position, years of experience, education level, or performance rating, which may also influence salary.
